### Import Liberaries

```python
import cv2
import numpy as np
from keras.models import model_from_json
```

### labeling our output

```python
emotion_dict = {0:"Angry", 1: "Disgusted" , 2:"Fearful", 3:"Happy", 4 :"Neutral", 5:"Sad", 6:"Surprised"}
```

### Load json abd create model
- **load the structure of the CNN model we built eariler**

```python
json_file =open("model/emotion_model.json", "r")

loaded_model_json = json_file.read()

json_file.close()
emotion_model = model_from_json(loaded_model_json)
```

### Load weights into new model
- **load the weights that our model learned in the training**

```python
emotion_model.load_weights("model/emotion_model.h5")
print("weights loaded sucessfully")
```

```python
# cap = cv2.VideoCapture(0)

cap = cv2.VideoCapture(r"sample/Example of Human Facial Expressions _ Emotions.mp4")
while True:
    # find hear cascade to draw bounding box 
    ret , frame = cap.read()
    frame = cv2.resize(frame , (1280 , 720))
    
    if not ret:
        break
    # use cascade calssifier to detect faces
    face_detector = cv2.CascadeClassifier(r"haarcascades/haarcascade_frontalface_default.xml")
    
    # convert our frame to gray
    gray_frame = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    
    # detect multi faces
    num_faces = face_detector.detectMultiScale(gray_frame, scaleFactor =1.3 , minNeighbors = 5)
    
    # preprocessing of the faces (detect coordinates of the face and draw a rectangle around it)
    for (x, y, w, h) in num_faces:
        cv2.rectangle(frame ,(x , y-50) ,(x+w , y+h+10), (0 ,255, 0), 4)
        roi_gray_frame = gray_frame[y:y + h, x: x + w]
        cropped_img = np.expand_dims(np.expand_dims(cv2.resize(roi_gray_frame, (48 , 48)) , -1), 0)
        
        # predict the emotions
        emotion_prediction = emotion_model.predict(cropped_img)
        # get the heighest probabilty of the emotion list [0 to 6]
        maxindex = int(np.argmax(emotion_prediction))
        # pass the index to emotion dictionary
        cv2.putText(frame , emotion_dict[maxindex], (x+5 , y-20),cv2.FONT_HERSHEY_SIMPLEX, 1, (255 , 0 , 0),2, cv2.LINE_AA)
        
    cv2.imshow("Emotion Detection" , frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindow()
    
```
